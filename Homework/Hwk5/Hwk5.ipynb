```python
# Exercise 6
import boto3
import pandas as pd; pd.set_option('display.max_columns', 100)
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import recall_score
from sklearn.ensemble import RandomForestClassifier

s3 = boto3.resource('s3')
bucket_name = 'data-445-timlincoln'
bucket = s3.Bucket(bucket_name)

file_key = 'framingham.csv'

bucket_object = bucket.Object(file_key)
file_object = bucket_object.get()
file_content_stream = file_object.get('Body')

# Reading the data-file
heart = pd.read_csv(file_content_stream)
heart.head()
```

```python
# Dropping N/A variables
heart = heart.dropna()
```

```python
# Defining the input and target variables
X = heart.loc[:, heart.columns != 'TenYearCHD']
Y = heart['TenYearCHD']
```

```python
RF_list[]

for n in range(0, 100):
    # Splitting the data
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
    
    # Building the random forest model
    RF_md = RandomForestClassifier(n_estimators = 500, max_depth = 3).fit(X_train, Y_train)

    # Getting importance
    importance = RF_md.coef_

    # Summary
    for i,v in enumerate(importance):
        e = (i, v)
        
    RF_list.append(e)
    
# Defining average function
def call_average(num):
    sum_num = 0
    for t in num:
        sum_num = sum_num + t
        
    avg = sum_num / len(num)
    return avg

print('The top five importance features are:', call_average(RF_list))
```

```python
md1_list = []
md2_list = []
md3_list = []

# Defining the input and target variables
X = heart[['age', 'currentSmoker', 'totChol', 'BMI', 'heartRate']]
Y = heart['TenYearCHD']

for m in range(0, 100):
    # Splitting the data
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
    
    # Building the 1st random forest model
    RF1_md = RandomForestClassifier(n_estimators = 500, max_depth = 3).fit(X_train, Y_train)
    
    # Predicting on test
    RF1_pred = RF1_md.predict_proba(X_test)[:, 1]

    # Changing to labels
    RF1_labels = np.where(RF1_pred < 0.1, 0, 1)
    
    g = recall_score(Y_test, RF1_labels)
    md1_list.append(g)
    
    # Building the 2nd random forest model
    RF2_md = RandomForestClassifier(n_estimators = 500, max_depth = 5).fit(X_train, Y_train)
    
    # Predicting on test
    RF2_pred = RF2_md.predict_proba(X_test)[:, 1]

    # Changing to labels
    RF2_labels = np.where(RF2_pred < 0.1, 0, 1)
    
    g = recall_score(Y_test, RF2_labels)
    md2_list.append(g)
    
    # Building the 3rd random forest model
    RF3_md = RandomForestClassifier(n_estimators = 500, max_depth = 7).fit(X_train, Y_train)
    
    # Predicting on test
    RF3_pred = RF3_md.predict_proba(X_test)[:, 1]

    # Changing to labels
    RF3_labels = np.where(RF3_pred < 0.1, 0, 1)
    
    g = recall_score(Y_test, RF3_labels)
    md3_list.append(g)
```

```python
# Defining average function
def call_average(num):
    sum_num = 0
    for t in num:
        sum_num = sum_num + t
        
    avg = sum_num / len(num)
    return avg

# Calling average function
print('Average recall for model 1 is:', call_average(md1_list))
print('Average recall for model 2 is:', call_average(md2_list))
print('Average recall for model 3 is:', call_average(md3_list))
```

```python
# From the results the best model for predicting TenYearCHD would be model 2 because of the highest recall score 
# meaning setting the max depth to 5 is just right for how deep the trees will go. 
```
